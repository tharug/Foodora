```python
import html5lib
import requests
from bs4 import BeautifulSoup
import json
import csv
```

```python
domain='https://www.foodora.com.au'
city='melbourne'
url=domain+'/city/'+city
req=requests.get(url)
```

```python
soup=BeautifulSoup(req.content,'lxml')
```

```python
main=soup.main.find_all('ul',{'class':'vendor-list closed '})
restaurants=main[0].find_all('a')
```

```python
vendor_list=[]
with open('food.csv','w') as csv_file:
    writer=csv.DictWriter(csv_file,fieldnames=['domain','city','name','address','price_point','cuisine','restaurant_url','delivery_hrs'])
    
    for restaurant in restaurants:
        vendor={}
        page=restaurant['href']
        vendor_details=json.loads(restaurant['data-vendor'])

        vendor['domain']=domain
        vendor['name']=vendor_details['name']
        vendor['city']=vendor_details['city']['name']
        address=vendor_details['address']
        address_line2=vendor_details['address_line2']
        locality=vendor_details['city']['name']
        post_code=vendor_details['post_code']
        vendor['address']=v_address(address,address_line2,post_code,locality)
        budget=vendor_details['budget']
        vendor['price_point']=price_point(budget)
        vendor['restaurant_url']=vendor_details['web_path']
        cuisine=set()

        for item in vendor_details['characteristics']['cuisines']:
            cuisine.add(item['name'])

        for item in vendor_details['characteristics']['food_characteristics']:
            cuisine.add(item['name'])
        
        primary_cuisine=vendor_details['characteristics']['primary_cuisine']['name']
        cuisine.add(primary_cuisine)

        vendor['cuisine']=','.join(cuisine)
        vendor['delivery_hrs']=get_delivery_hrs(vendor_details['web_path'])
        
        writer.writerow(vendor)
        print (vendor['restaurant_url'])
        vendor_list.append(vendor)
```

```python
def v_address(address,address_line2,post_code,city):
    if address_line2 is None:
        address_line2=','
    else:
        address_line2=','+address_line2+','
    return address+address_line2+str(post_code)+' '+city
```
